```python
def rid_sol(protein_mol2_file_path):
    """
    Given a path to a `protein.mol2` file, removes the `HOH` and `GTP` lines and returns the modified contents as a string.

    Parameters:
    -----------
    protein_mol2_file_path : str
        The path to the `protein.mol2` file.

    Returns:
    --------
    str
        The modified contents of the `protein.mol2` file as a string.
    """
    with open(protein_mol2_file_path, 'r') as f:
        lines = f.readlines()

    new_lines = []
    for line in lines:
        if line.startswith('HETATM') and 'HOH' not in line and 'GTP' not in line:
            new_lines.append(line)

    return ''.join(new_lines)
```

```python
import openbabel
import os
from concurrent.futures import ThreadPoolExecutor

def convert_mol2_to_pdb(mol2_file_path: str, pdb_file_path: str):
    """
    Convert a MOL2 file to a PDB file using Open Babel.

    Parameters:
    -----------
    mol2_file_path : str
        The path to the input MOL2 file.
    pdb_file_path : str
        The path to the output PDB file.
    """
    # Create Open Babel objects
    ob_conversion = openbabel.OBConversion()
    ob_mol = openbabel.OBMol()

    # Set input and output formats
    ob_conversion.SetInAndOutFormats("mol2", "pdb")

    # Read the MOL2 file
    ob_conversion.ReadFile(ob_mol, mol2_file_path)

    # Write the PDB file
    ob_conversion.WriteFile(ob_mol, pdb_file_path)


def convert_mol2_to_pdb_wrapper(pdb_id, input_path, output_path):
    mol2_file_path = os.path.join(input_path, pdb_id, 'protein.mol2')
    pdb_file_path = os.path.join(output_path, pdb_id, 'protein.pdb')
    convert_mol2_to_pdb(mol2_file_path, pdb_file_path)

# Example usage
input_path = './bench2'
output_path = './bench2'

if not os.path.exists(output_path):
    os.makedirs(output_path)

pdb_ids = [name for name in os.listdir(input_path) if os.path.isdir(os.path.join(input_path, name))]

print(pdb_ids)

# Set the number of threads to the number of available CPU cores
num_threads = os.cpu_count()

with ThreadPoolExecutor(max_workers=num_threads) as executor:
    executor.map(convert_mol2_to_pdb_wrapper, pdb_ids, [input_path] * len(pdb_ids), [output_path] * len(pdb_ids))
```

```python
import os
import pandas as pd


def calculate_ligand_center(mol2_file_path):
    """
    Given a path to a `ligand.mol2` file, calculates and returns the center coordinates of the ligand.

    Parameters:
    -----------
    mol2_file_path : str
        The path to the `ligand.mol2` file.

    Returns:
    --------
    tuple of floats
        The x, y, and z coordinates of the center of the ligand.
    """
    with open(mol2_file_path, 'r') as f:
        lines = f.readlines()

    lines = [line.split() for line in lines]

    x_sum, y_sum, z_sum, n_atoms = 0, 0, 0, 0

    for line in lines:
        if len(line) == 9:
            x, y, z, *_ = line[2:5]
            x_sum += float(x)
            y_sum += float(y)
            z_sum += float(z)
            n_atoms += 1

    center_x, center_y, center_z = x_sum/n_atoms, y_sum/n_atoms, z_sum/n_atoms
    return center_x, center_y, center_z
```

```python
import os
import pandas as pd
import numpy as np
from typing import Tuple, List

non_standard_to_standard = {
    '2AS':'ASP', '3AH':'HIS', '5HP':'GLU', 'ACL':'ARG', 'AGM':'ARG', 'AIB':'ALA', 'ALM':'ALA', 'ALO':'THR', 'ALY':'LYS', 'ARM':'ARG',
    'ASA':'ASP', 'ASB':'ASP', 'ASK':'ASP', 'ASL':'ASP', 'ASQ':'ASP', 'ASX':'ASP', 'AYA':'ALA', 'BCS':'CYS', 'BHD':'ASP', 'BMT':'THR', 'BNN':'ALA', # Added ASX => ASP
    'BUC':'CYS', 'BUG':'LEU', 'C5C':'CYS', 'C6C':'CYS', 'CAS':'CYS', 'CCS':'CYS', 'CEA':'CYS', 'CGU':'GLU', 'CHG':'ALA', 'CLE':'LEU', 'CME':'CYS',
    'CSD':'ALA', 'CSO':'CYS', 'CSP':'CYS', 'CSS':'CYS', 'CSW':'CYS', 'CSX':'CYS', 'CXM':'MET', 'CY1':'CYS', 'CY3':'CYS', 'CYG':'CYS',
    'CYM':'CYS', 'CYQ':'CYS', 'DAH':'PHE', 'DAL':'ALA', 'DAR':'ARG', 'DAS':'ASP', 'DCY':'CYS', 'DGL':'GLU', 'DGN':'GLN', 'DHA':'ALA',
    'DHI':'HIS', 'DIL':'ILE', 'DIV':'VAL', 'DLE':'LEU', 'DLY':'LYS', 'DNP':'ALA', 'DPN':'PHE', 'DPR':'PRO', 'DSN':'SER', 'DSP':'ASP',
    'DTH':'THR', 'DTR':'TRP', 'DTY':'TYR', 'DVA':'VAL', 'EFC':'CYS', 'FLA':'ALA', 'FME':'MET', 'GGL':'GLU', 'GL3':'GLY', 'GLZ':'GLY',
    'GMA':'GLU', 'GSC':'GLY', 'HAC':'ALA', 'HAR':'ARG', 'HIC':'HIS', 'HIP':'HIS', 'HMR':'ARG', 'HPQ':'PHE', 'HTR':'TRP', 'HYP':'PRO',
    'IAS':'ASP', 'IIL':'ILE', 'IYR':'TYR', 'KCX':'LYS', 'LLP':'LYS', 'LLY':'LYS', 'LTR':'TRP', 'LYM':'LYS', 'LYZ':'LYS', 'MAA':'ALA', 'MEN':'ASN',
    'MHS':'HIS', 'MIS':'SER', 'MLE':'LEU', 'MPQ':'GLY', 'MSA':'GLY', 'MSE':'MET', 'MVA':'VAL', 'NEM':'HIS', 'NEP':'HIS', 'NLE':'LEU',
    'NLN':'LEU', 'NLP':'LEU', 'NMC':'GLY', 'OAS':'SER', 'OCS':'CYS', 'OMT':'MET', 'PAQ':'TYR', 'PCA':'GLU', 'PEC':'CYS', 'PHI':'PHE',
    'PHL':'PHE', 'PR3':'CYS', 'PRR':'ALA', 'PTR':'TYR', 'PYL':'LYS', 'PYX':'CYS', 'SAC':'SER', 'SAR':'GLY', 'SCH':'CYS', 'SCS':'CYS', 'SCY':'CYS', 'SEC':'CYS', # Added pyrrolysine and selenocysteine
    'SEL':'SER', 'SEP':'SER', 'SET':'SER', 'SHC':'CYS', 'SHR':'LYS', 'SMC':'CYS', 'SOC':'CYS', 'STY':'TYR', 'SVA':'SER', 'TIH':'ALA',
    'TPL':'TRP', 'TPO':'THR', 'TPQ':'ALA', 'TRG':'LYS', 'TRO':'TRP', 'TYB':'TYR', 'TYI':'TYR', 'TYQ':'TYR', 'TYS':'TYR', 'TYY':'TYR'
}

fa_letter = {
    'ALA': 'A', 'ARG': 'R', 'ASN': 'N', 'ASP': 'D', 'CYS': 'C',
    'GLN': 'Q', 'GLU': 'E', 'GLY': 'G', 'HIS': 'H', 'ILE': 'I',
    'LEU': 'L', 'LYS': 'K', 'MET': 'M', 'PHE': 'F', 'PRO': 'P',
    'SER': 'S', 'THR': 'T', 'TRP': 'W', 'TYR': 'Y', 'VAL': 'V',
    'UNK': 'X'}

def extract_AA(file_path: str) -> str:
    amino_acids = []
    residue_numbers = set()
    residue_last_CA = {}

    with open(file_path, 'r') as mol2_file:
        lines = mol2_file.readlines()
            
        for line in lines:
            if line.startswith('@<TRIPOS>BOND'):
                break

                
            split_line = line.split()
            if len(split_line) >= 7:  # Check if the line has enough items
                residue_number = int(split_line[6])
                if residue_number not in residue_numbers:
                    residue_numbers.add(residue_number)
                    residue_last_CA[residue_number] = ''
                    three_letter_code = split_line[7][:3]
                    one_letter_code = fa_letter.get(three_letter_code, 'X')
                    amino_acids.append(one_letter_code)
                atom_name = split_line[1]
                if atom_name == 'CA':
                    residue_last_CA[residue_number] = one_letter_code

        # Get the last CA atom for each residue
        for residue_number, aa in residue_last_CA.items():
            if aa:
                amino_acids.append(aa)

    return ''.join(amino_acids)



def labeling(file_path: str, center_coordinates: List[float], threshold: float = 10.0) -> Tuple[str, str]:
    labels = []
    coords = []
    center_coordinates_array = np.array(center_coordinates)

    with open(file_path, 'r') as mol2_file:
        lines = mol2_file.readlines()

        for line in lines:
            if line.startswith('@<TRIPOS>BOND'):
                break

            split_line = line.split()
            if len(split_line) >= 7:
                atom_name = split_line[1]

                if atom_name == 'CA':
                    coord = np.array([float(split_line[2]), float(split_line[3]), float(split_line[4])])
                    distance = np.linalg.norm(coord - center_coordinates_array)

                    if distance <= threshold:
                        labels.append('1')
                        coords.append(coord)
                    else:
                        labels.append('0')
    if len(coord) == 0:
        print(f"Warning: No coordinates found for CA atoms in {file_path}")
        return '', ''
    
    label_coord = np.mean(coords, axis=0)
    #print(label_coord)
    label_coord = f"{label_coord[0]}, {label_coord[1]}, {label_coord[2]}"

    return ''.join(labels), label_coord


# Read data.tsv and extract pdb_ids
df = pd.read_csv('data.tsv', sep='\t')
pdb_ids = df['pdb']
pdb_ids_list = list(pdb_ids)

df['AA'] = ''
df['labels'] = ''
df['CCA'] = ''

input_path = "./data"

for pdb_id in pdb_ids:
    pdb_file_path = os.path.join(input_path, pdb_id, f'protein.mol2')
    mol2_file_path = os.path.join(input_path, pdb_id, f'ligand.mol2')

    if not os.path.exists(pdb_file_path):
        print(f"Bad input for {pdb_id}: protein.mol2 file not found.")
    elif not os.path.exists(mol2_file_path):
        print(f"Bad input for {pdb_id}: ligand.mol2 file not found.")
    else:
        amino_acids = extract_AA(pdb_file_path)
        df.loc[df['pdb'] == pdb_id, 'AA'] = ''.join(amino_acids)

        center = calculate_ligand_center(mol2_file_path)

        labels, label_coord = labeling(pdb_file_path, center, threshold=10.0)
        df.loc[df['pdb'] == pdb_id, 'labels'] = ''.join(str(label) for label in labels)
        
        df.loc[df['pdb'] == pdb_id, 'CCA'] = str(label_coord)


df.to_csv('data.tsv', index=False, sep='\t')
```

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

def get_cath_data(pdb_id):
    """
    pdb_id를 인자로 받아 CATH database에서 단백질 구조 데이터를 가져온 후, 이를 이용하여 노드의 특성을 나타내는 원핫벡터를 생성한다.
    """
    # CATH database에서 단백질 구조 데이터 가져오기
    response = requests.get(f'https://www.cathdb.info/pdb/{pdb_id}/cath_summary')
    if response.status_code != 200:
        raise ValueError(f'CANNOT retrieve CATH data for PDB ID: {pdb_id}')
    soup = BeautifulSoup(response.content, 'html.parser')

    # CATH 분류, 아키텍처, 토폴로지 정보 추출
    cath_classification = None
    cath_architecture = None
    cath_topology = None
    cath_classification_elem = soup.select('#cath-classification .summary-value')
    cath_architecture_elem = soup.select('#cath-architecture .summary-value')
    cath_topology_elem = soup.select('#cath-topology .summary-value')

    if cath_classification_elem:
        cath_classification = cath_classification_elem[0].text.strip()
    if cath_architecture_elem:
        cath_architecture = cath_architecture_elem[0].text.strip()
    if cath_topology_elem:
        cath_topology = cath_topology_elem[0].text.strip()

    # CATH 분류, 아키텍처, 토폴로지 정보를 원핫인코딩하여 벡터로 변환
    cath_classification_vec = [int(cath_classification == f'1.{i}.0') for i in range(1, 16)]
    cath_architecture_vec = [int(cath_architecture == f'1.10.{i}.10') for i in range(1, 21)]
    cath_topology_vec = [int(cath_topology == f'1.10.10.{i}') for i in range(1, 78)]

    # CATH 데이터를 하나의 벡터로 합쳐서 반환
    cath_data_vec = cath_classification_vec + cath_architecture_vec + cath_topology_vec
    return cath_data_vec


# 데이터 불러오기
df = pd.read_csv('data.tsv', sep='\t')
pdb_ids = df['pdb']
pdb_ids_list = list(pdb_ids)

# pdb_ids_list에 있는 모든 pdb 아이디에 대해 CATH 데이터 가져오기
cath_data_list = []
for pdb_id in pdb_ids_list:
    try:
        cath_data = get_cath_data(pdb_id)
        cath_data_list.append(cath_data)
    except ValueError as e:
        print(str(e))

print(f'Total number of PDB IDs: {len(pdb_ids_list)}')
print(f'Number of PDB IDs with CATH data: {len(cath_data_list)}')
```

```python
import requests
from bs4 import BeautifulSoup

def fetch_ecod_info(pdb_id):
    """
    Fetch ECOD information for a given PDB ID.

    :param pdb_id: str, PDB ID to search for.
    :return: List of tuples, containing (PDB ID, ECOD ID) pairs.
    """
    # Access the search result page on the ECOD website
    url = f'http://prodata.swmed.edu/ecod/complete/search?kw={pdb_id}&type=pdbid'
    response = requests.get(url)
    html = response.content

    # Parse the search result page
    soup = BeautifulSoup(html, 'html.parser')
    tbody = soup.find('tbody')

    # Extract and print ECOD information
    ecod_info = []
    if tbody is not None:
        for tr in tbody.find_all('tr'):
            cols = tr.find_all('td')
            ecod_info.append((cols[0].text, cols[1].text))

    return ecod_info
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from tqdm import tqdm
import concurrent.futures

num_cores = 8

# 데이터 로드 및 PDB ID 추출
df = pd.read_csv('data.tsv', sep='\t')
pdb_ids = df['pdb']
pdb_ids_list = list(pdb_ids)

# 결과를 저장할 빈 DataFrame 생성
ecod_data = pd.DataFrame(columns=['PDB ID', 'ECOD ID'])

# 멀티코어 프로세싱을 사용하여 각 PDB ID에 대해 검색하고 결과를 DataFrame에 추가
with concurrent.futures.ThreadPoolExecutor(max_workers=num_cores) as executor:
    results = list(tqdm(executor.map(fetch_ecod_info, pdb_ids_list), total=len(pdb_ids_list)))

# 결과를 DataFrame에 추가
for pdb_id, ecod_info in zip(pdb_ids_list, results):
    for pdb, ecod in ecod_info:
        ecod_data = ecod_data.append({'PDB ID': pdb, 'ECOD ID': ecod}, ignore_index=True)

# 결과 DataFrame 출력 및 CSV 파일로 저장
print(ecod_data)
ecod_data.to_csv('ecod_data.csv', index=False)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from tqdm import tqdm
import concurrent.futures
import os

def fetch_ecod_info(pdb_id):
    """
    Fetch ECOD information for a given PDB ID.

    :param pdb_id: str, PDB ID to search for.
    :return: List of tuples, containing (domain ID, range) pairs.
    """
    # Access the search result page on the ECOD website
    url = f'http://prodata.swmed.edu/ecod/complete/search?kw={pdb_id}&type=pdbid'
    response = requests.get(url)
    html = response.content

    # Parse the search result page
    soup = BeautifulSoup(html, 'html.parser')
    tbody = soup.find('tbody')

    # Extract and return ECOD information
    ecod_info = []
    if tbody is not None:
        for tr in tbody.find_all('tr'):
            cols = tr.find_all('td')
            ecod_info.append((cols[1].text, cols[0].text))

    return ecod_info

# Load data and extract PDB IDs
df = pd.read_csv('data.tsv', sep='\t')
pdb_ids = df['pdb']
pdb_ids_list = list(pdb_ids)


# Create an empty DataFrame to store the results
ecod_data = pd.DataFrame(columns=['PDB ID', 'Domain ID', 'Range'])

# Check the number of available CPU cores
num_cores = 8
print(f"Using {num_cores} cores")

# Search for each PDB ID using multicore processing and add the results to the DataFrame in the correct order
with concurrent.futures.ThreadPoolExecutor(max_workers=num_cores) as executor:
    # Map each PDB ID to a future
    futures = {executor.submit(fetch_ecod_info, pdb_id): pdb_id for pdb_id in pdb_ids_list}
    
    for future in tqdm(concurrent.futures.as_completed(futures), total=len(pdb_ids_list)):
        pdb_id = futures[future]
        ecod_info = future.result()
        for domain_id, pdb_range in ecod_info:
            ecod_data = ecod_data.append({'PDB ID': pdb_id, 'Domain ID': domain_id, 'Range': pdb_range}, ignore_index=True)

# Print and save the resulting DataFrame
print(ecod_data)
ecod_data.to_csv('ecod_data.csv', index=False)
```

SCOP 데이터베이스에서 사용하는 도메인 ID는 7자리로 구성되어 있습니다. 각 자리는 다음과 같은 의미를 가지고 있습니다.

Class: 단백질 도메인이 속하는 구조적 클래스(Class)를 나타냅니다.
Fold: 단백질 도메인이 속하는 구조적 Fold를 나타냅니다.
Superfamily: 단백질 도메인이 속하는 구조적 Superfamily를 나타냅니다.
Family: 단백질 도메인이 속하는 구조적 Family를 나타냅니다.
Protein: 단백질 도메인이 속하는 단백질을 나타냅니다.
Species: 해당 단백질 도메인이 발견된 생물종을 나타냅니다.
Domain: 도메인 ID에서 가장 마지막 자리를 의미하며, 하나의 단백질에 여러 개의 도메인이 존재할 경우, 각 도메인에 대해 서로 다른 도메인 ID를 부여하기 위해 사용됩니다.
따라서, SCOP에서는 도메인 ID의 앞 3자리가 Fold를 나타내는 것입니다. 예를 들어, "d1tfga_"는 Fold가 "TIM beta/alpha-barrel"인 단백질 도메인을 나타냅니다.

```python
def check_CA_coords(file_path: str):
    with open(file_path, 'r') as mol2_file:
        lines = mol2_file.readlines()

        for i, line in enumerate(lines):
            split_line = line.split()
            if len(split_line) >= 7:
                atom_name = split_line[1]

                if atom_name == 'CA':
                    try:
                        coord = (float(split_line[2]), float(split_line[3]), float(split_line[4]))
                        print(f"Line {i+1}: {coord}")
                    except ValueError:
                        print(f"Warning: No coordinates found for CA atom on line {i+1}")

file_path = "./coach420/3m4e/protein.mol2"
check_CA_coords(file_path)
```
